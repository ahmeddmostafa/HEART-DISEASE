```python
import streamlit as st
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
```

```python
# Load trained model
model = joblib.load("heart_disease_model.pkl")
```

```python
# Streamlit App
st.title("Heart Disease Prediction App")
st.write("Enter your health details below to predict the risk of heart disease.")
```

```python
# User Inputs
age = st.number_input("Age", 20, 100, 50)
sex = st.selectbox("Sex", ["Male", "Female"])
cp = st.selectbox("Chest Pain Type (cp)", [0,1,2,3])
trestbps = st.number_input("Resting Blood Pressure", 80, 200, 120)
chol = st.number_input("Serum Cholesterol (mg/dl)", 100, 600, 200)
fbs = st.selectbox("Fasting Blood Sugar > 120 mg/dl (fbs)", [0,1])
restecg = st.selectbox("Resting ECG Results", [0,1,2])
thalach = st.number_input("Max Heart Rate Achieved", 70, 250, 150)
exang = st.selectbox("Exercise Induced Angina", [0,1])
oldpeak = st.number_input("ST Depression (oldpeak)", 0.0, 6.0, 1.0, step=0.1)
slope = st.selectbox("Slope of ST segment", [0,1,2])
ca = st.selectbox("Number of Major Vessels (ca)", [0,1,2,3,4])
thal = st.selectbox("Thalassemia (thal)", [0,1,2,3])

# Convert input to model format
sex_val = 1 if sex == "Male" else 0
input_data = np.array([[age, sex_val, cp, trestbps, chol, fbs, restecg,
                        thalach, exang, oldpeak, slope, ca, thal]])

# Prediction
if st.button("Predict"):
    prediction = model.predict(input_data)[0]
    if prediction == 1:
        st.error("⚠️ The model predicts that you **may have Heart Disease**.")
    else:
        st.success("✅ The model predicts that you **do not have Heart Disease**.")
```

```python
# Data Visualization (Bonus)
st.subheader(" Heart Disease Dataset Overview")

df = pd.read_csv("processed.cleveland.data", names=[
    "age","sex","cp","trestbps","chol","fbs","restecg",
    "thalach","exang","oldpeak","slope","ca","thal","target"
], na_values="?")
df = df.dropna()
df["target_binary"] = (df["target"] > 0).astype(int)

st.write("Sample of dataset:", df.head())

# Plot
fig, ax = plt.subplots()
df["age"].hist(bins=20, ax=ax)
ax.set_title("Age Distribution in Dataset")
ax.set_xlabel("Age")
ax.set_ylabel("Count")
st.pyplot(fig)

fig2, ax2 = plt.subplots()
df.groupby("target_binary")["chol"].mean().plot(kind="bar", ax=ax2)
ax2.set_title("Average Cholesterol by Heart Disease Presence")
ax2.set_xlabel("Heart Disease (0=No, 1=Yes)")
ax2.set_ylabel("Avg Cholesterol")
st.pyplot(fig2)
```
